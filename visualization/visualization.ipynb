```python
import os
import pickle

import matplotlib.pyplot as plt
```

```python
with open("D:\Programming\DP\models\System-Imbalance-Forecasting\data\data_TS\merged_data.pkl", 'rb') as file:
    ds_raw =  pickle.load(file)
```

```python
data = ds_raw[1000:1100]
```

```python
time = data.datetime_utc
cont1 = data.system_imbalance
cont2 = data.total_load
cont3 = data.hour_sin
cont4 = data['measured_&_upscaled_solar']

# Reverting categorical plots to scatter plots
fig, axs = plt.subplots(4, 1, figsize=(8, 4), sharex=True)

# Line plots for continuous variables
axs[0].plot(time, cont1, alpha=0.7)
axs[1].plot(time, cont2, alpha=0.7)
axs[2].plot(time, cont3, alpha=0.7)
axs[3].plot(time, cont4, alpha=0.7)

axs[0].spines['bottom'].set_visible(False)
for ax in axs[1:3]:
    for spine in ax.spines.values():
        spine.set_linestyle('--')
axs[3].spines['top'].set_linestyle('--')

# Making the plot minimalistic
plt.subplots_adjust(hspace=0)
plt.setp(axs, xticks=[], yticks=[])
plt.show()
```
