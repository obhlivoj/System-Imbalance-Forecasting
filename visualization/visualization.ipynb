```python
# Activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

plt.rcParams['font.size'] = 12
plt.rcParams['axes.labelsize'] = 14
plt.rcParams['axes.titlesize'] = 16
plt.rcParams['xtick.labelsize'] = 12
plt.rcParams['ytick.labelsize'] = 12
plt.rcParams['legend.fontsize'] = 12

# Generate a range of values
x = np.linspace(-10, 10, 100)

# Compute the activation functions
y_sigmoid = sigmoid(x)
y_tanh = tanh(x)
y_relu = relu(x)

# Create plots
plt.figure(figsize=(16, 4))

# Sigmoid plot
plt.subplot(1, 3, 1)
plt.plot(x, y_sigmoid, label="Sigmoid", color="blue", linewidth=3)
#plt.title("Sigmoid")
plt.xlabel("x")
plt.ylabel("sigmoid(x)")
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
#plt.legend()

# Tanh plot
plt.subplot(1, 3, 2)
plt.plot(x, y_tanh, label="Tanh", color="red", linewidth=3)
#plt.title("Tanh")
plt.xlabel("x")
plt.ylabel("tanh(x)")
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
#plt.legend()

# ReLU plot
plt.subplot(1, 3, 3)
plt.plot(x, y_relu, label="ReLU", color="green", linewidth=3)
#plt.title("ReLU")
plt.xlabel("x")
plt.ylabel("ReLU(x)")
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
#plt.legend()

# Adjust layout
# Display only the axes

plt.tight_layout()
plt.savefig("activation.pdf", format="pdf", bbox_inches="tight")
plt.show()
```
